```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv('movies.csv', encoding= 'unicode_escape')
```

```python
df
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
print(df.dtypes)
```

```python
df.boxplot(column=['gross'])
matplotlib.rcParams['figure.figsize'] = (7,8)
```

```python
df.drop_duplicates()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
sns.regplot(x="gross", y="budget", data=df)
matplotlib.rcParams['figure.figsize'] = (15,8)
```

```python
sns.regplot(x="score", y="gross", data=df)
```

```python
df['votes'] = df['votes'].fillna(0)  # Fill NaN with 0
df['votes'] = df['votes'].replace([np.inf, -np.inf], 0)  # Replace inf values with 0
df['votes'] = df['votes'].astype(int)

df['runtime'] = df['runtime'].fillna(0)  # Fill NaN with 0
df['runtime'] = df['runtime'].replace([np.inf, -np.inf], 0)  # Replace inf values with 0
df['runtime'] = df['runtime'].astype(int)
```

```python
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr(method='pearson')
print(correlation_matrix)
```

```python
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr(method='kendall')
print(correlation_matrix)
```

```python
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr(method='spearman')
print(correlation_matrix)
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
